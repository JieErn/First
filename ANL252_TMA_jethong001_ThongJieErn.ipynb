```python
#Engagement survey outcomes/rating

import pandas as pd
import matplotlib.pyplot as plt
import statistics

Survey_outcomes = pd.read_csv("TMA_Data_Original.csv")
Engagement_rating = list(Survey_outcomes["Survey"])

rating_min = min(Engagement_rating)
rating_max = max(Engagement_rating)
rating_median = statistics.median(Engagement_rating)
rating_mean = statistics.mean(Engagement_rating)

print(f"the minimum rating is {rating_min}")
print(f"the maximum rating is {rating_max}")
print(f"the median rating is {rating_median}")
print(f"the mean rating is {rating_mean}")

plt.figure(figsize=(5,5), dpi=80)

plt.hist(Engagement_rating, bins=4, range=(rating_min, rating_max), align="mid", orientation="vertical", rwidth=0.9)
plt.title("Engagement Rating")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()


```
